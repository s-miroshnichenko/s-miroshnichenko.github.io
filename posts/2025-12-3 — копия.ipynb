## Объяснение нейросетей на примере

### Задача

Возьмем задачу классификации текстов "I like cats" and "I like dogs". Мы хотим понять какой текст про собак, а какой - про кошек. 

### Словарь

| Token index| Token |
|--------|---------|
| 1    | I      |
| 2  | like      |
| 3 | cats      |
|4|dogs|


### Репрезентация токенов

Token "I" is represented with all elements set to 0, except 1.

| Token index| Token |One-hot encoded vector|
|--------|---------|-|
| 1    | I      |1
| 2  | like      |0
| 3 | cats      |0
|4|dogs|0


"I" token также можно репрезентовать как вектор [1], a "I like cats" как сумму векторов. Тогда получился [1, 2, 3]. 

### Input layer

Это вектор индексов [1, 2, 3]. 

### Embedding слой

```python
# Инициализация эмбеддинг-слоя
embeds = nn.Embedding(vocab_size = 4, embedding_dim = 3)

# Передача индексов "I like cats"
input_indicies = torch.tensor([1, 2, 3])
embeddings = embeds(input_indicies)
```

В результате мы получим матрицу размерности $ (\text{кол-во слов}) \times \text{embedding\_dim} $.


Размерность эмбеддинга выбирают исходя из модели и сложности задачи. В нашем примере на каджое слово будет приходиться 3 признака. 

#### Пример результата

```python
tensor([[0.35, -0.21, 0.76],    # embedding for "I"
        [-0.7, 0.675, -0.8],    # embedding for "like"
        [0.66, -0.33, 0.32]])   # embedding for "cats"
```
