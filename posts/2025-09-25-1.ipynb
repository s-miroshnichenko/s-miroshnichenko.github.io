title: "Почему я завел этот блог"
date: 2025-09-25
---
Я нашел классный [курс по генеративному ии](https://www.edx.org/certificates/professional-certificate/ibm-generative-ai-engineering?index=product&queryId=43fd79a173e724e5b5abad1e34712be1&position=1). Его можно проходить бесплатно.

Завел блог по рекомендации Jeremy Howard. Он является автором еще одного [крутого бесплатного курса](https://course.fast.ai/) по глубокому обучению. Я благодарен ему за одну очень ценную рекомендацию по ораганизации процесса обучения: после прохождения урока стирать код в ноутбуке (под ноутбуками в этом блоге я буду подразумевать Jupyter notebook files) и писать его самому. Выполнение этой рекомендации невероятно обогащает процесс обучения, потому что когда просто читаешь код: "А, все понятно", а когда пишешь задаешься вопросами: "почему нужно вызывать вызывать эту функцию и как ее найти, а вот у нее еще есть параметр, а что будет если запускать ее без него?"

Еще во мне дремлет учительская душа. Мне так хочется обучать (и общаться с юношами), когда я нахожу какой-то классный курс. Может и пойду проведу бесплатный курс в колледже.

> Еще один плюс ведения блога. Изначально я написал что это желание разбивается о невозможности прилично на этом заработать, но пока я писал эти строки, у меня появились идея что можно обкатать свой курс на живой аудитории, а потом разместить его на udemy.

Еще одна причина завести блог кроется в моем желании накидать костяк для этого курса.
